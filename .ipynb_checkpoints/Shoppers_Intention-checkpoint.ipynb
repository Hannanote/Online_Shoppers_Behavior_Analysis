```python
month_dict = {'Jan':1,'Feb':2,'Mar':3, 'Apr':4, 'May':5, 'June':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
sns.countplot(x = X['Month'], data = y, hue = "Revenue", order = month_dict.keys())
plt.show()
```

```python
print(X['Weekend'].value_counts())
print(y.value_counts())

data = pd.concat([X,y],axis = 1)
sns.countplot(x='Weekend', data=data, hue= 'Revenue')
plt.show()
```

```python
# Categorical data
sns.countplot(x = X['VisitorType'], data = y, hue = 'Revenue')
plt.show()
```

```python
sns.countplot(x = X.Month, data = X, hue = X.VisitorType, order = month_dict.keys())
plt.show()
```

```python
sns.countplot(x=X['Month'], hue = X['VisitorType'], data = data)
plt.show()
```

```python
mon_rev = X.groupby('Month')['VisitorType'].value_counts().unstack().fillna(0)
print(mon_rev)

month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
mon_rev.index = pd.Categorical(mon_rev.index, categories=month_order, ordered=True)
mon_rev = mon_rev.sort_index()
print(mon_rev)

# Plot the results
mon_rev.plot(kind='line')
plt.show()
```
